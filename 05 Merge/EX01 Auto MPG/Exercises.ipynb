### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the first dataset [cars1](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv) and [cars2](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv).  

```python
data_cars01 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv')
data_cars02 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv')
```

```python
data_cars01.head(10)
```

```python
data_cars02.head(10)
```

### Step 3. Oops, it seems our first dataset has some unnamed blank columns, fix cars1

```python
data_cars01.info()
```

```python
data_cars01 = data_cars01.dropna(axis='columns')
data_cars01.head(10)
```

### Step 4. What is the number of observations in each dataset?

```python
data_cars01.shape[0]
```

```python
data_cars02.shape[0]
```

### Step 5. Join cars1 and cars2 into a single DataFrame called cars

```python
dataset = pd.concat([data_cars01, data_cars02])
dataset.shape[0]
```

### Step 6. Oops, there is a column missing, called owners. Create a random number Series from 15,000 to 73,000.

```python
dataset['owners'] = np.random.randint(low=15000, high=73001, size=dataset.shape[0])
```
